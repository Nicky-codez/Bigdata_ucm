```python
import matplotlib.pyplot as plt
import numpy as np

# Create the same plot as above with 2 rows and 2 columns and figsize of (10, 5)
fig, axs = plt.subplots(2, 2, figsize=(10, 5))

# Plot x versus x/2 on the top left axes
x = np.linspace(0, 100, 100)
axs[0, 0].plot(x, x / 2, color='royalblue')
axs[0, 0].set_title('x vs x/2')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('x/2')

# Plot a scatter plot of 10 random numbers on each axis on the top right subplot
x_rand = np.random.rand(10)
y_rand = np.random.rand(10)
axs[0, 1].scatter(x_rand, y_rand, color='coral')
axs[0, 1].set_title('Random Scatter Plot')
axs[0, 1].set_xlabel('x values')
axs[0, 1].set_ylabel('y values')

# Plot a bar graph of the favourite food keys and values on the bottom left subplot
favorite_foods = {
    'Vada': 12.99,
    'Masala Dosa': 8.75,
    'Idly': 15.50
}
food_names = list(favorite_foods.keys())
food_prices = list(favorite_foods.values())
axs[1, 0].bar(food_names, food_prices, color='seagreen')
axs[1, 0].set_title('Favorite Foods')
axs[1, 0].set_xlabel('Food Item')
axs[1, 0].set_ylabel('Price (USD)')

# Plot a histogram of 1000 random normally distributed numbers on the bottom right subplot
x_hist = np.random.randn(1000)
axs[1, 1].hist(x_hist, bins=30, color='plum', edgecolor='black')
axs[1, 1].set_title('Histogram of Normal Data')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
car_sales = pd.read_csv('car-sales.csv')

# Clean the 'Price' column by removing symbols and trimming the final two digits
car_sales['Price'] = (
    car_sales['Price']
    .replace(r'[\$,]', '', regex=True)
    .str[:-2]
    .astype(float)
)

# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column
car_sales['Total Sales'] = car_sales['Price'].cumsum()

# Add a column called 'Sale Date' which lists a series of successive dates starting from today
car_sales['Sale Date'] = pd.date_range(start=pd.Timestamp.today().normalize(), periods=len(car_sales), freq='D')

# View the car_sales DataFrame
car_sales
```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
car_sales.plot(x='Sale Date', y='Total Sales', kind='line')

plt.title('Cumulative Car Sales Over Time')
plt.xlabel('Sale Date')
plt.ylabel('Total Sales (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Wonderful, you've changed the style of the plots and the figure is looking different but the dots aren't a very good colour.

Let's change the `cmap` parameter of `scatter()` as well as the `color` parameter of `axhline()` to fix it.

Completing this step correctly should result in a figure which looks like plot3

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
heart_disease = pd.read_csv('heart-disease.csv')

# Filter for patients over 50
over_50 = heart_disease[heart_disease['age'] > 50]

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data using a scatter plot with cmap='winter'
scatter = ax.scatter(
    over_50.index,
    over_50['age'],
    c=over_50['age'],           # Color by age values
    cmap='winter',
    edgecolor='black'
)

# Customize the plot
ax.set_title('Age Distribution of Patients Over 50')
ax.set_xlabel('Patient Index')
ax.set_ylabel('Age')

# Add a meanline
mean_age = over_50['age'].mean()
ax.axhline(mean_age, color='red', linestyle='--', linewidth=2, label=f'Mean Age: {mean_age:.1f}')
ax.legend()

# Add colorbar for the scatter
plt.colorbar(scatter, ax=ax, label='Age (Color Scale)')
plt.tight_layout()
plt.show()
```
